```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import necessary libraries
import os
import zipfile
import pretty_midi
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define the path to the ZIP file on Google Drive
zip_path = '/content/drive/My Drive/AAI511_FinalProject-main.zip'
extract_path = '/content/drive/My Drive/AAI511_FinalProject'

# Extract the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Verify the extracted contents
for root, dirs, files in os.walk(extract_path):
    print(f"Root: {root}")
    for name in dirs:
        print(f"Dir: {os.path.join(root, name)}")
    for name in files:
        print(f"File: {os.path.join(root, name)}")

# Adjust the base directory based on the verified structure
base_dir = os.path.join(extract_path, 'AAI511_FinalProject-main/Composer_Dataset/NN_midi_files_extended/train')

# Check if the base directory exists
if not os.path.exists(base_dir):
    print(f"Base directory does not exist: {base_dir}")
else:
    # List of folders (composers) within the base directory
    composer_folders = os.listdir(base_dir)

    # Initialize lists to store extracted features
    all_note_events = []
    all_velocities = []
    all_durations = []
    all_instruments = []
    all_composers = []

    # Loop through each composer folder
    for composer_folder in composer_folders:
        composer_path = os.path.join(base_dir, composer_folder)

        # Check if the item in the base directory is a folder
        if os.path.isdir(composer_path):
            print(f"Loading and processing MIDI files from {composer_folder}...")

            # Loop through each MIDI file in the composer folder
            for file_name in os.listdir(composer_path):
                file_path = os.path.join(composer_path, file_name)

                # Check if the item in the composer folder is a file and has a .mid or .midi extension
                if os.path.isfile(file_path) and (file_name.endswith('.mid') or file_name.endswith('.midi')):
                    try:
                        midi_data = pretty_midi.PrettyMIDI(file_path)

                        # Extract note events, velocities, durations, and instruments
                        note_events = []
                        velocities = []
                        durations = []
                        instruments = []

                        for instrument in midi_data.instruments:
                            for note in instrument.notes:
                                note_events.append(note.pitch)
                                velocities.append(note.velocity)
                                durations.append(note.end - note.start)  # Duration in seconds
                                instruments.append(instrument.program)  # MIDI program number

                        all_note_events.append(note_events)
                        all_velocities.append(velocities)
                        all_durations.append(durations)
                        all_instruments.append(instruments)

                        # Extract composer (assuming composer name is before the first underscore)
                        composer = file_name.split('_')[0]
                        all_composers.append(composer)

                        print(f" Loaded and processed: {file_name}")
                    except Exception as e:
                        print(f" Error processing {file_name}: {str(e)}")

    print("MIDI data processing complete!")

    # Pad sequences to ensure uniform length
    all_note_events_padded = pad_sequences(all_note_events, padding='post', value=-1)
    all_velocities_padded = pad_sequences(all_velocities, padding='post', value=-1)
    all_durations_padded = pad_sequences(all_durations, padding='post', value=-1)
    all_instruments_padded = pad_sequences(all_instruments, padding='post', value=-1)

    # Convert lists to NumPy arrays for further processing
    all_note_events = np.array(all_note_events_padded)
    all_velocities = np.array(all_velocities_padded)
    all_durations = np.array(all_durations_padded)
    all_instruments = np.array(all_instruments_padded)
    all_composers = np.array(all_composers)

    # Example: Print shapes of extracted features
    print("Shapes of extracted features:")
    print("Note Events:", all_note_events.shape)
    print("Velocities:", all_velocities.shape)
    print("Durations:", all_durations.shape)
    print("Instruments:", all_instruments.shape)
    print("Composers:", all_composers.shape)

```

```python
import os
import pretty_midi
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Path to the directory containing MIDI files
base_dir = '/content/drive/My Drive/AAI511_FinalProject/AAI511_FinalProject-main/Composer_Dataset/NN_midi_files_extended/train'

# List of folders (composers) within the base directory
composer_folders = os.listdir(base_dir)

# Initialize lists to store extracted features and labels
x_training_data = []
y_training_data = []

# Encoder for composer labels
label_encoder = LabelEncoder()

# Loop through each composer folder
for composer_folder in composer_folders:
    composer_path = os.path.join(base_dir, composer_folder)

    # Check if the item in the base directory is a folder
    if os.path.isdir(composer_path):
        print(f"Loading MIDI files from {composer_folder}...")

        # Loop through each MIDI file in the composer folder
        for file_name in os.listdir(composer_path):
            file_path = os.path.join(composer_path, file_name)

            # Check if the item in the composer folder is a file and has a .mid or .midi extension
            if os.path.isfile(file_path) and (file_name.endswith('.mid') or file_name.endswith('.midi')):
                try:
                    midi_data = pretty_midi.PrettyMIDI(file_path)

                    # Extract musical features (example: note events)
                    note_events = []
                    for instrument in midi_data.instruments:
                        for note in instrument.notes:
                            note_events.append(note.pitch)

                    # Append to x_training_data (assuming fixed sequence length)
                    # Adjust sequence length based on your needs
                    sequence_length = 100  # Example sequence length
                    if len(note_events) >= sequence_length:
                        x_training_data.append(note_events[:sequence_length])
                        y_training_data.append(composer_folder)  # Use folder name as label

                    print(f" Loaded: {file_name}")
                except Exception as e:
                    print(f" Error loading {file_name}: {str(e)}")

# Convert lists to numpy arrays
x_training_data = np.array(x_training_data)
y_training_data = np.array(y_training_data)

# Encode composer labels
y_training_data_encoded = label_encoder.fit_transform(y_training_data)

# Pad sequences to ensure uniform length
x_training_data_padded = pad_sequences(x_training_data, padding='post', value=-1)

# Example: Print shapes of extracted features and labels
print("Shapes:")
print("x_training_data:", x_training_data_padded.shape)
print("y_training_data_encoded:", y_training_data_encoded.shape)




```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Normalize the note events
x_training_data_normalized = x_training_data_padded / 127.0  # MIDI note range is 0-127

# One-hot encode the labels
onehot_encoder = OneHotEncoder(sparse=False)
y_training_data_onehot = onehot_encoder.fit_transform(y_training_data_encoded.reshape(-1, 1))

# Print shapes to verify
print("x_training_data_normalized shape:", x_training_data_normalized.shape)
print("y_training_data_onehot shape:", y_training_data_onehot.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

# Define the CNN model
model = Sequential()
model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(x_training_data_normalized.shape[1], 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.2))
model.add(Conv1D(128, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(y_training_data_onehot.shape[1], activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
# Reshape x_training_data_normalized for CNN input
x_training_data_normalized = np.expand_dims(x_training_data_normalized, axis=-1)

# Train the model
history = model.fit(x_training_data_normalized, y_training_data_onehot, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Assuming new_data_normalized is preprocessed similarly to x_training_data_normalized
# new_data_normalized = ...

# Example: Predict on the training data (for demonstration purposes)
predictions = model.predict(x_training_data_normalized)

# Decode the predictions
predicted_labels = onehot_encoder.inverse_transform(predictions)

# Print example predictions
print(predicted_labels[:10])
```

```python
from sklearn.metrics import accuracy_score, f1_score

# Evaluate the model on the training data
train_loss, train_accuracy = model.evaluate(x_training_data_normalized, y_training_data_onehot, verbose=0)
print(f'Training Accuracy: {train_accuracy:.4f}')

# Make predictions on the training data
y_train_pred = model.predict(x_training_data_normalized)
y_train_pred_labels = np.argmax(y_train_pred, axis=1)

# Decode the one-hot encoded training labels
y_train_true_labels = np.argmax(y_training_data_onehot, axis=1)

# Calculate accuracy and F1 score for the training data
train_accuracy = accuracy_score(y_train_true_labels, y_train_pred_labels)
train_f1_score = f1_score(y_train_true_labels, y_train_pred_labels, average='weighted')

print(f'Training Accuracy: {train_accuracy:.4f}')
print(f'Training F1 Score: {train_f1_score:.4f}')
```

```python
# Evaluate the model on the validation data
validation_loss, validation_accuracy = model.evaluate(x_training_data_normalized, y_training_data_onehot, verbose=0)
print(f'Validation Accuracy: {validation_accuracy:.4f}')

# Make predictions on the validation data
y_val_pred = model.predict(x_training_data_normalized)
y_val_pred_labels = np.argmax(y_val_pred, axis=1)

# Decode the one-hot encoded validation labels
y_val_true_labels = np.argmax(y_training_data_onehot, axis=1)

# Calculate accuracy and F1 score for the validation data
val_accuracy = accuracy_score(y_val_true_labels, y_val_pred_labels)
val_f1_score = f1_score(y_val_true_labels, y_val_pred_labels, average='weighted')

print(f'Validation Accuracy: {val_accuracy:.4f}')
print(f'Validation F1 Score: {val_f1_score:.4f}')
```
